```python
import pandas as pd
dataset = pd.read_csv('ks-projects-201801.csv')
dataset.head()
```

```python
dataset.info()
```

```python
dataset.launched = pd.to_datetime(dataset.launched)
dataset.deadline = pd.to_datetime(dataset.deadline)
dataset.info()
```

```python
dataset['duration'] = dataset.deadline - dataset.launched
```

```python
dataset['duration'] = dataset['duration'].apply(lambda x: x.days)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
def success (x) :
    if  x >= 1 :
        return 0
    else :
        return 1
```

```python
# if dataset.loc[dataset['goal'] < dataset['pledged']]:
#     dataset['success'] = 1
# else :
#     dataset['success'] = 0
```

```python
dataset['success'] = dataset.goal - dataset.pledged
```

```python
dataset.head()
```

```python
dataset['success'] = dataset['success'].apply(success)
dataset.head()
```

```python
dataset = dataset.drop(columns=['ID','name','currency','deadline','launched','goal','pledged','state','country','usd pledged','usd_pledged_real','usd_goal_real'])
dataset
```

```python
dataset.isna().sum()
dataset.info()
```

```python
x = dataset.iloc[:,:-1]
x
```

```python
x= pd.get_dummies(x)
x
```

```python
y = dataset.iloc[:,-1]
y
```

```python
# # standrize the values
# from sklearn.preprocessing import StandardScaler , MinMaxScaler
# scaler = MinMaxScaler()
# x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split ,GridSearchCV
x_train ,x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2 ,random_state=0)
```

```python
# from sklearn.neighbors import KNeighborsClassifier 
# cl = KNeighborsClassifier(n_neighbors = 22)
# cl.fit(x_train,y_train)
# from sklearn.linear_model import LogisticRegression
# cl = LogisticRegression(random_state = 0)
# cl.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier , plot_tree
dtclf = DecisionTreeClassifier(max_depth = 8, min_samples_leaf = 7)
dtclf.fit(x_train,y_train)
```

```python
parameters = [{'min_samples_leaf':[2,5,7],'max_depth':[5,7,8]}]
grid_search = GridSearchCV(estimator = dtclf , param_grid = parameters , cv=10 , n_jobs = -1)
grid_search = grid_search.fit(x_train, y_train)
print("best accuracy is :" , grid_search.best_score_)
grid_search.best_params_
```

```python
y_pred = dtclf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix ,f1_score , precision_score ,recall_score ,classification_report
cm = confusion_matrix(y_test, y_pred)
print (pd.DataFrame(cm))
print(classification_report(y_test, y_pred))
```
